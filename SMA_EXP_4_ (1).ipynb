```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'dataset.csv' with your dataset file path)
data = pd.read_csv('/content/expt4 dataset sales.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head(20))

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# Summary statistics of numerical columns
print("\nSummary statistics of numerical columns:")
print(data.describe())


```

```python

# Histograms for numerical columns
data.hist(figsize=(10, 8))
plt.suptitle('Histograms')
plt.show()


```

```python
# Box plots for numerical columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Box plots')
plt.xticks(rotation=45)
plt.show()

```

```python
# Pairplot for visualizing relationships between numerical columns
sns.pairplot(data)
plt.title('Pairplot')
plt.show()

```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Bar plot for categorical columns
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='DEALSIZE')
plt.title('Count of Categorical Data ')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot for categorical columns
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='COUNTRY')
plt.title('Count of Categorical Data ')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(25,6 ))
sns.barplot(data=data, x='CITY', y='SALES')
plt.title('Bar Graph between CITY and SALES')
plt.xlabel('CITY')
plt.ylabel('SALES')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
