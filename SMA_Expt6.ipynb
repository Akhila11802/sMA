```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv('Friendship - Friendship-network_data_2013.csv')
```

```python
df.head()
```

```python
fb_graph = nx.from_pandas_edgelist(df, source=df.columns[0], target=df.columns[1])
print(type(fb_graph))
```

```python
pos = nx.spring_layout(fb_graph)
plt.figure(figsize=(10, 8))  # You can adjust the values as needed

# Draw the graph with labels
nx.draw(fb_graph, pos=pos, with_labels=True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(6, 6))  # Smaller figure size for more zoomed-in view

# Draw the graph with labels, increasing the distance between nodes
nx.draw(fb_graph, with_labels=True, node_size=1000, node_color='skyblue', edge_color='gray', linewidths=1, font_size=10)

# Show the plot
plt.show()
```

```python

# Generate a 3D layout for the graph
pos = nx.spring_layout(fb_graph, dim=3)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Draw the graph with labels
nx.draw(fb_graph, pos=pos, with_labels=True, node_size=100, ax=ax)

# Show the plot
plt.show()
```

```python
# pos = nx.spring_layout(fb_graph)  # Experiment with different layout algorithms
pos = nx.spring_layout(fb_graph, k=0.15)  # Increase k value to increase node separation
nx.draw(fb_graph, pos=pos, with_labels=True, node_size=100, node_color='skyblue', edge_color='gray')
```

```python
import matplotlib.pyplot as plt
```

```python
pos = nx.spring_layout(G)
betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)
node_color = [20000.0 * G.degree(v) for v in G]
node_size = [v* 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G, pos=pos, with_labels=False, node_color=node_color, node_size=node_size)
#plt.axis('off)
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

```python
closeness_centrality= nx.centrality.closeness_centrality(G)
(sorted(closeness_centrality.items(), key = lambda item: item[1], reverse=True))[:8]
```

```python
node_size = [
    v * 50 for v in closeness_centrality.values()
]
plt.figure(figsize=(15,8))
nx.draw_networkx(G, pos=pos, node_size = node_size, with_labels = False, width=0.15)
plt.axis("off")
```

```python
bridges = list(nx.bridges(G))
```

```python
len(bridges)
```

```python
local_bridges = list(nx.local_bridges(G, with_span = False))
len(local_bridges)
```

```python
plt.figure(figsize=(15,8))
nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15)
nx.draw_networkx_edges(
    G, pos, edgelist=local_bridges, width=0.5, edge_color="red"
)

nx.draw_networkx_edges(
    G, pos, edgelist=bridges, width=0.7, edge_color="yellow"
)
# plt.axis("off")
```
