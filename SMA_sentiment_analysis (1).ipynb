```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

```

```python
# Text preprocessing function
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    # Join tokens back into a single string
    preprocessed_text = ' '.join(lemmatized_tokens)

    return preprocessed_text

# Apply text preprocessing to 'description' and 'title' columns
df_cleaned['description'] = df_cleaned['description'].apply(preprocess_text)
df_cleaned['title'] = df_cleaned['title'].apply(preprocess_text)
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer


# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Perform sentiment analysis on 'description' and 'title' columns
df_cleaned['description_sentiment'] = df_cleaned['description'].apply(lambda x: sid.polarity_scores(x)['compound'])
df_cleaned['title_sentiment'] = df_cleaned['title'].apply(lambda x: sid.polarity_scores(x)['compound'])

# Display the DataFrame with sentiment scores
print(df_cleaned[['description', 'description_sentiment', 'title', 'title_sentiment']])
```

```python
def get_sentiment_label(score):
    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"

# Apply the function to the sentiment scores
df_cleaned['description_sentiment_label'] = df_cleaned['description_sentiment'].apply(get_sentiment_label)
df_cleaned['title_sentiment_label'] = df_cleaned['title_sentiment'].apply(get_sentiment_label)

# Display the DataFrame with sentiment labels
print(df_cleaned[['description', 'description_sentiment_label', 'title', 'title_sentiment_label']])
```

```python
# Drop 'title', 'title_sentiment', and 'likecount' columns
df_cleaned.drop(['title', 'title_sentiment', 'title_sentiment_label'], axis=1, inplace=True)

# Display the DataFrame with dropped columns
print(df_cleaned)
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Function to perform sentiment analysis using TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to the 'description' column
df_cleaned['description_sentiment'] = df_cleaned['description'].apply(analyze_sentiment)
```

```python
# Count the number of positive, negative, and neutral sentiments
sentiment_counts = df_cleaned['description_sentiment'].value_counts()

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
